# Select the features uding FPR method with alpa 0.01 an if class

```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectFpr, f_classif
```

```python
x, y = load_iris(return_X_y=True)
var=SelectFpr(score_func=f_classif, alpha=0.05).fit_transform(x,y)

print(var.shape)
```

# Create a plot using matplotlin of state london and paris display as triangle

```python
import matplotlib.pyplot as plt
import numpy as np

x=1
y=np.sin(x)
z=np.cos(x)
plt.plot(x, y, label='london', color='blue', marker='^', markersize=20)
plt.plot(x, z, label='paris', color='red', marker='^', markersize=20)
plt.title('London vs Paris')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
