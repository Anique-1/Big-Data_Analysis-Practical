#  NumPy (Numerical Python) 

### create a zeros array with given shape and datatype

```python
import numpy as np
# Create a zeros

zeros = np.zeros([2,3])

print(zeros)
```

### create ones array with given datatype

```python
var=np.ones([3,5,6,8], dtype=np.int64)
var
```

### Empty array creation

```python
np.empty([2,4,6,8])
```

### create Arrange array

```python
arrange=np.arange(1,10,2)
arrange
```

### create linspace array

```python
linespace=np.linspace(0,10,6100)
linespace
```

### How to Sort the array

```python
arr=[1, 4, 2, 5, 3, 6, 8, 9, 7, 10]
np.sort(arr, kind='quicksort')
```

### argsort array index

```python
np.argsort(arr)
```

```python
### Generate an empty 4 dimensional array fill elements in it using for loop

arr_4d = np.empty((4, 4, 4, 4))
for i in range(4):
    for j in range(4):
        for k in range(4):
            for l in range(4):
                arr_4d[i, j, k, l] = i * 1000 + j * 100 + k * 10 + l
                print(arr_4d)
```

```python
# identity function
arr_identity = np.identity(3)
print(arr_identity)
```
